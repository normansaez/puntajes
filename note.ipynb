```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('data/puntajes.csv')
```

```python
df
```

```python
df = df[df['numero de heat 1'] != 1]
df = df[df['numero de heat 1'] != 4]
df = df[df['numero de heat 1'] != 7]
df = df[df['numero de heat 1'] != 10]
df = df[df['numero de heat 1'] != 13]
```

```python
df = df[df['numero de heat 2'] != 16]
df = df[df['numero de heat 2'] != 19]
df = df[df['numero de heat 2'] != 22]
```

```python
df
```

```python
 df = df.fillna(0)
```

```python
df
```

```python
df['OLA1_OLA2'] = df['OLA1'] + df['OLA2']
df['OLA3_OLA4'] = df['OLA3'] + df['OLA4']

df['puesto heat 1'] = 0
df['puesto heat 2'] = 0
df
```

```python
for i in range(2,12+1):
    heat = df[df['numero de heat 1'] == i]
    heat['puesto heat 1'] = heat['OLA1_OLA2'].rank(method='min', ascending=False).astype(int)
    df.loc[heat.index, 'puesto heat 1'] = heat['puesto heat 1'].values
    if i == 3:
        print(heat)
#    print(df_seleccionado)
```

```python
for i in range(12,24+1):
    heat = df[df['numero de heat 2'] == i]
    heat['puesto heat 2'] = heat['OLA3_OLA4'].rank(method='min', ascending=False).astype(int)
    df.loc[heat.index, 'puesto heat 2'] = heat['puesto heat 2'].values
```

```python
df['puntaje_p1']= 0
df['puntaje_p2']= 0
```

```python
df
```

```python
def multiplicar_por_condicion(row):
    if row['puesto heat 1'] == 1:
        row['puntaje_p1']=10
        return row['puntaje_p1']
    elif row['puesto heat 1'] == 2:
        row['puntaje_p1']= 8.6
        return row['puntaje_p1']
    elif row['puesto heat 1'] == 3:
        row['puntaje_p1'] = 7.3
        return row['puntaje_p1']
    elif row['puesto heat 1'] == 4:
        row['puntaje_p1'] = 6.7
        return row['puntaje_p1']    
    else:
        return row['puntaje_p1']

# Aplicar la función a cada fila del DataFrame y guardar el resultado en una nueva columna 'D'
df['puntaje_p1'] = df.apply(multiplicar_por_condicion, axis=1)

print(df)
```

```python
df
```

```python
def multiplicar_por_condicion(row):
    if row['puesto heat 2'] == 1:
        row['puntaje_p2']=10
        return row['puntaje_p2']
    elif row['puesto heat 2'] == 2:
        row['puntaje_p2']= 8.6
        return row['puntaje_p2']
    elif row['puesto heat 2'] == 3:
        row['puntaje_p2'] = 7.3
        return row['puntaje_p2']
    elif row['puesto heat 2'] == 4:
        row['puntaje_p2'] = 6.7
        return row['puntaje_p2']    
    else:
        return row['puntaje_p2']

# Aplicar la función a cada fila del DataFrame y guardar el resultado en una nueva columna 'D'
df['puntaje_p2'] = df.apply(multiplicar_por_condicion, axis=1)

print(df)
```

```python
df.to_csv('output.csv', index=False)
```

```python
df['PuestoP'] = df['puntaje_p1']+df['puntaje_p2']+df['Sumatoria de Olas']
```

```python
df_ordenado = df.sort_values(by='PuestoP', ascending=False)
```

```python
df_ordenado['PuestoPP'] = df_ordenado['PuestoP'].rank(method='min', ascending=False).astype(int)
```

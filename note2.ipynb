```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv('data/competidoras.csv', sep=';')
```

```python
df[['OLA1', 'OLA2','OLA3', 'OLA4','SUMATORIA','ola1_ola2','ola3_ola4']] = df[['OLA1', 'OLA2','OLA3', 'OLA4','SUMATORIA','ola1_ola2','ola3_ola4']].apply(lambda x: x.str.replace(',', '.')).astype(float) 
```

```python
df['puesto heat 1'] = 0
df['puesto heat 2'] = 0
df
```

```python
for i in range(1,12+1):
    heat = df[df['heat1'] == i]
    heat['puesto heat 1'] = heat['ola1_ola2'].rank(method='min', ascending=False).astype(int)
    df.loc[heat.index, 'puesto heat 1'] = heat['puesto heat 1'].values
    if i == 3:
        print(heat)
df
```

```python
for i in range(12,24+1):
    heat = df[df['heat2'] == i]
    heat['puesto heat 2'] = heat['ola3_ola4'].rank(method='min', ascending=False).astype(int)
    df.loc[heat.index, 'puesto heat 2'] = heat['puesto heat 2'].values
```

```python
df
```

```python
df['puntaje_p1']= 0
df['puntaje_p2']= 0
```

```python
df
```

```python
def multiplicar_por_condicion(row):
    if row['puesto heat 1'] == 1:
        row['puntaje_p1']=10
        return row['puntaje_p1']
    elif row['puesto heat 1'] == 2:
        row['puntaje_p1']= 8.6
        return row['puntaje_p1']
    elif row['puesto heat 1'] == 3:
        row['puntaje_p1'] = 7.3
        return row['puntaje_p1']
    elif row['puesto heat 1'] == 4:
        row['puntaje_p1'] = 6.7
        return row['puntaje_p1']    
    else:
        return row['puntaje_p1']

# Aplicar la funci√≥n a cada fila del DataFrame y guardar el resultado en una nueva columna 'D'
df['puntaje_p1'] = df.apply(multiplicar_por_condicion, axis=1)

print(df)
```

```python
df
```

```python
def multiplicar_por_condicion(row):
    if row['puesto heat 2'] == 1:
        row['puntaje_p2']=10
        return row['puntaje_p2']
    elif row['puesto heat 2'] == 2:
        row['puntaje_p2']= 8.6
        return row['puntaje_p2']
    elif row['puesto heat 2'] == 3:
        row['puntaje_p2'] = 7.3
        return row['puntaje_p2']
    elif row['puesto heat 2'] == 4:
        row['puntaje_p2'] = 6.7
        return row['puntaje_p2']    
    else:
        return row['puntaje_p2']

# Aplicar la funci√≥n a cada fila del DataFrame y guardar el resultado en una nueva columna 'D'
df['puntaje_p2'] = df.apply(multiplicar_por_condicion, axis=1)

print(df)
```

```python
#df[['OLA1', 'OLA2','OLA3', 'OLA4']] = df[['OLA1', 'OLA2','OLA3', 'OLA4']].apply(lambda x: x.str.replace(',', '.')).astype(float)
```

```python
df['Sumatoria de Olas'] = df['OLA1'].astype(float)+df['OLA2'].astype(float) +df['OLA3'].astype(float) + df['OLA4'].astype(float)
```

```python
df['PuestoP'] = df['puntaje_p1']+df['puntaje_p2']+df['Sumatoria de Olas']
```

```python
df_ordenado = df.sort_values(by='PuestoP', ascending=False)
```

```python
df_ordenado['PuestoPP'] = df_ordenado['PuestoP'].rank(method='min', ascending=False).astype(int)
df_ordenado['Puesto'] = df_ordenado['SUMATORIA'].rank(method='min', ascending=False).astype(int)
```

```python
df_ordenado
```

```python
df_f = df_ordenado[['NOMBRE COMPETIDORA','Puesto','PuestoPP']]
df_f = df_f.reset_index(drop=True)
```

```python
df_ordenado.to_csv('data/completo.csv', index=False)
df_f.to_csv('data/final.csv', index=False)
```
